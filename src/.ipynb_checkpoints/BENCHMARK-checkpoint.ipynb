### Xception model - gradCAM1

```python
model_builder = keras.applications.xception.Xception
img_size = (299, 299)
preprocess_input = keras.applications.xception.preprocess_input
decode_predictions = keras.applications.xception.decode_predictions
```

```python
last_conv_layer_name = "block14_sepconv2_act"
classifier_layer_names = [
    "avg_pool",
    "predictions",
]

# The local path to our target image
# img_path = keras.utils.get_file(
#     "african_elephant.jpg", " https://i.imgur.com/Bvro0YD.png"
# )
```

```python
img_path="./cat_dog.PNG"
display(Image(img_path))
```

```python
# Prepare image
img_array = preprocess_input(get_img_array(img_path, size=img_size))

# Make model
model = model_builder(weights="imagenet")

# Print what the top predicted class is
preds = model.predict(img_array)
print("Predicted:", decode_predictions(preds, top=1)[0])
model.summary()
```

```python
# Generate class activation heatmap
heatmap = make_gradcam_heatmap(
    img_array, model, last_conv_layer_name, classifier_layer_names
)

# Display heatmap
plt.matshow(heatmap)
plt.show()
```

```python
"""
## Create a superimposed visualization
"""

# We load the original image
img = keras.preprocessing.image.load_img(img_path)
img = keras.preprocessing.image.img_to_array(img)

# We rescale heatmap to a range 0-255
heatmap = np.uint8(255 * heatmap)

# We use jet colormap to colorize heatmap
jet = cm.get_cmap("jet")

# We use RGB values of the colormap
jet_colors = jet(np.arange(256))[:, :3]
jet_heatmap = jet_colors[heatmap]

# We create an image with RGB colorized heatmap
jet_heatmap = keras.preprocessing.image.array_to_img(jet_heatmap)
jet_heatmap = jet_heatmap.resize((img.shape[1], img.shape[0]))
jet_heatmap = keras.preprocessing.image.img_to_array(jet_heatmap)

# Superimpose the heatmap on original image
superimposed_img = jet_heatmap * 0.4 + img
superimposed_img = keras.preprocessing.image.array_to_img(superimposed_img)
```

```python
# Save the superimposed image
save_path = "./cat_dog_cam.jpg"
superimposed_img.save(save_path)
```

```python
# Display Grad CAM
display(Image(save_path))
```

### Inception

```python
from keras.applications import inception_v3
inet_model = inception_v3.InceptionV3()
inet_model
```

```python
def transform_img_inception (path_list):
    out = []
    for path in path_list:
        img = image.load_img(path, target_size=(299, 299))
        x = image.img_to_array(img)
        x = np.expand_dims(x, axis=0)
        x = inception_v3.preprocess_input(x)
        out.append(x)
    return np.vstack(out)
```

```python
images_inc = transform_img_inception([os.path.join('cat_dog.png')])
# I'm dividing by 2 and adding 0.5 because of how this Inception represents images
plt.imshow(images_inc[0] / 2 + 0.5)
```

```python
preds = inet_model.predict(images_inc)
for x in decode_predictions(preds)[0]:
    print(x)
```
